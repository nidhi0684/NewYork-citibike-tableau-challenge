```python
# Dependencies and Setup
import pandas as pd
```

```python
# List of file names to be combined 
files22 = ['data/202201.csv', 'data/202202.csv', 'data/202203.csv', 'data/202204.csv', 'data/202205.csv', 'data/202206.csv', 'data/202207.csv', 'data/202208.csv', 'data/202209.csv', 'data/202210.csv', 'data/202211.csv', 'data/202212.csv']

# Create an empty list to store each dataframe
df22_list = []

# Loop through each file in the list
for file in files22:
    # Read the file into a dataframe
    df22 = pd.read_csv(file, parse_dates=['started_at','ended_at'])

    # Append the dataframe to the list
    df22_list.append(df22)

# Concatenate all dataframes in the list into one dataframe
df22 = pd.concat(df22_list)

#Add new column 'trip_duration' based on 'started_at' & 'ended_at' columns
df22['trip_duration'] = (df22['ended_at'] - df22['started_at']).dt.seconds

# Show dataframe
df22
```

```python
# List of file names to be combined 
files23 = ['data/202301.csv', 'data/202302.csv', 'data/202303.csv', 'data/202304.csv', 'data/202305.csv', 'data/202306.csv', 'data/202307.csv', 'data/202308.csv', 'data/202309.csv', 'data/202310.csv', 'data/202311.csv', 'data/202312.csv']

# Create an empty list to store each dataframe
df23_list = []

# Loop through each file in the list
for file in files23:
    # Read the file into a dataframe
    df23 = pd.read_csv(file, parse_dates=['started_at','ended_at'])

    # Append the dataframe to the list
    df23_list.append(df23)

# Concatenate all dataframes in the list into one dataframe
df23 = pd.concat(df23_list)

#Add new column 'trip_duration' based on 'started_at' & 'ended_at' columns
df23['trip_duration'] = (df23['ended_at'] - df23['started_at']).dt.seconds

# Show dataframe
df23
```

```python
# Concatenate all dataframes for all the years into one
combined_df = pd.concat([df22, df23], axis=0)

# Display the results
combined_df
```

```python
clean_df=combined_df.dropna(subset=['end_station_name', 'start_station_name'])
```
